```python
from pathlib import Path

def find_csv_file_path(filename, directory='.'):
    search_path = Path(directory)
    for path in search_path.rglob(f'{filename}'):
        return str(path.resolve())
    return None

import pandas as pd
print("Enter the CSV filename (with .csv extension):")
df = pd.read_csv(find_csv_file_path("sample.csv"))
print("--- Orders DataFrame Loaded ---")
print(df.head())

print("Total Amounts of All Orders:")
print(df[['total_amount']].to_string(header=False))
print(f"Name: total_amount, dtype: {df['total_amount'].dtype}")

print("Details of the 4th order:")
print(df.iloc[3])
print(f"Average Total Amount: ₹{df['total_amount'].mean():.2f}")

print("Updated details for ORD0004:")
df.loc[3, "order_status"] = "Cancelled"
df.loc[3, "payment_method"] = "UPI"
print(df[df['order_id'] == 'ORD0004'])
```
